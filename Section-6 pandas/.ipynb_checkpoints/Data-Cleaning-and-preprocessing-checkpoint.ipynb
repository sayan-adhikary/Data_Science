```python
df2 = df.dropna().copy()
```

```python
df2["Age"] = df2["Age"].astype(int)
```

```python
df2
```

```python
df2.info
```

```python
df2["Age Group"] = df2["Age"].apply(lambda x: "Adult" if x >= 125 else "Minor")
```

```python
df2
```

```python
gender_map = {"M": "Male", "F": "Female"}
df2["Gender"] = df2["Gender"].map(gender_map)
```

```python
df2
```

```python
df2["City"] = df2["City"].replace({"Del": "Delhi", "Mum": "Mumbai"})
```
